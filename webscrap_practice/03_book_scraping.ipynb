1. 질문 : 검색어로 찾은 책 목록을 json 파일로 저장

```python
import requests
from dotenv import load_dotenv
import os
import json

load_dotenv()

NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

headers = {
    "X-Naver-Client-Id": NAVER_CLIENT_ID,
    "X-Naver-Client-Secret": NAVER_CLIENT_SECRET
}

def search_books(query):
    #quary string 문자열을 dict 선언
    payload ={
        'query': query,
        'display': 10,
        'sort': 'sim'
    }
    url= 'https://openapi.naver.com/v1/search/book.json'

    # requests get(url, params, headers) 요청 
    res = requests.get(url, params=payload, headers=headers)

    print(res.status_code)

    # data = res.json()
    # print(data)

    #json()함수로 응답 결과 가져오기
    items_data=res.json()['items']
    return items_data

books = search_books("파이썬 기초")

with open("books.json", "w", encoding="utf-8") as f:
    json.dump(books, f, ensure_ascii=False, indent=2)
```

2. books.json 파일을 pandas dataFrame로 저장
3. 질문 : 검색어로 찾은 책 목록 출력

```python
import pandas as pd

with open("books.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df
```

4. 질문 : 검색어로 찾은 책 목록 중에서 가격이 2만원 이상인 책만 출력

```python
import json
import pandas as pd

with open("books.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df["discount"] = pd.to_numeric(df["discount"], errors="coerce")

filtered_df = df[df["discount"] >= 20000]

filtered_df
```
