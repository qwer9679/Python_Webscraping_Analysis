```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
url = 'https://news.daum.net/society'

req_header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

res = requests.get(url, headers=req_header)
res.encoding = 'utf-8'
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')

    
    for li_tag in soup.select('ul.list_newsheadline2 li'):
        a_tag = li_tag.find('a')
        link = a_tag['href']	
            
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')
        title = strong_tag.text.strip()

        print(title)
        print(link)
        print('-'*40)
```

```python
section_dict = {
    '기후/환경':'climate',
    '사회':'society',
    '경제':'economy',
    '정치':'politics',
    '국제':'world',
    '문화':'culture',
    '생활':'life',
    'IT/과학':'tech',
    '인물':'people'
}

def print_news(section_name):
    url = f'https://news.daum.net/{section_dict[section_name]}'
    req_header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
    }

    res = requests.get(url, headers=req_header)
    res.encoding = 'utf-8'
    if res.ok:
        soup = BeautifulSoup(res.text, 'html.parser')

        print("[["+section_name+"]]")
        for li in soup.select('ul.list_newsheadline2 li'):
            a_tag = li.find('a')
            title_tag = li.select_one('strong.tit_txt')

            if not a_tag or not title_tag:
                continue
            
            print(title_tag.text.strip())
            print(a_tag['href'])
            print('-' * 40)
        print()

print_news('경제')
print_news('사회')
```
