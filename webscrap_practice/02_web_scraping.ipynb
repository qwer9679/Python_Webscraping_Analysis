```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image, display

section_dict = {
    '최신뉴스':'n0100',
    '정치':'n0200',
    '경제':'n0300',
    '사회':'n0400',
    '세계':'n0500',
    'IT/과학':'n0600'
}

def fix_img_url(url):
    if url.startswith('//'):
        return 'https:' + url
    return url


def print_news(section_name):
    url = f'https://news.nate.com/recent?mid={section_dict[section_name]}'
    headers = {
        'user-agent':'Mozilla/5.0'
    }

    res = requests.get(url, headers=headers)
    res.encoding = res.apparent_encoding

    if not res.ok:
        return

    soup = BeautifulSoup(res.text, 'html.parser')

    print(f'[[ {section_name} ]]')

    for li in soup.select('div.mduSubjectList div.mlt01')[:3]:
        a_tag = li.select_one('a')
        title_tag = li.select_one('h2.tit')
        img_tag = li.select_one('img')

        if not a_tag or not title_tag:
            continue

        title = title_tag.get_text(strip=True)
        link = 'https://news.nate.com' + a_tag.get('href', '')

        print(title)
        if img_tag:
            img_url = img_tag.get('src', '')
            img_url = fix_img_url(img_url)
            display(Image(url=img_url))
        print(link)

        print('-' * 40)
    print('-' * 40)


print_news('최신뉴스')
print_news('정치')
print_news('경제')
print_news('사회')
print_news('세계')
print_news('IT/과학')
```
